```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Book 8.xlsx")
data.head()
```

```python
dummies = pd.get_dummies(data.city)
dummies
```

```python
merged = pd.concat([ data,dummies ], axis=1)
merged
```

```python
final_data = merged.drop(['city','assam'],axis=1)
final_data.head()
```

```python
x= final_data.drop(['price'],axis=1)
#x.head()
y = final_data.price
```

```python
from sklearn.model_selection import  train_test_split
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.predict([[3000,0,0,1]])
```

```python
model.predict([[4000,1,0,0]])
```

## Using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle = data
dfle.city = le.fit_transform(dfle.city)
dfle
```

```python
x= dfle[['city','area']].values
x
```

```python
y = dfle.price.values
y
```

## Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('city', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:]
x
```

```python
model.fit(x,y)
```

```python
model.predict([[0,0,1,3400]])
```
